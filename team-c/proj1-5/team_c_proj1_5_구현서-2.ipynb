```python
# 앱 코드를 파일로 저장
%%writefile app.py
import pandas as pd
import numpy as np
import streamlit as st
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime
from google.cloud import bigquery
import os
os.environ["GOOGLE_CLOUD_PROJECT"] = "bike-analysis-project-452711"

# 페이지 설정
st.set_page_config(
    page_title="서울시 공공자전거(따릉이) 현황 대시보드",
    page_icon="🚲",
    layout="wide"
)

# 데이터 로드 함수
@st.cache_data(ttl=300)  # 5분마다 데이터 갱신
def load_data():
    # BigQuery 클라이언트 초기화
    client = bigquery.Client()

    # 최신 데이터 쿼리 (최신 시간의 데이터만 가져오기)
    query = """
    WITH latest AS (
      SELECT MAX(collection_time) as latest_time
      FROM `bike-analysis-project-452711.bike_analysis_project.seoul_bike_stations`
    )
    SELECT *
    FROM `bike-analysis-project-452711.bike_analysis_project.seoul_bike_stations`
    WHERE collection_time = (SELECT latest_time FROM latest)
    """

    # 쿼리 실행 및 DataFrame으로 변환
    df = client.query(query).to_dataframe()

    # 날짜/시간 컬럼 처리
    df['collection_time'] = pd.to_datetime(df['collection_time'])

    # 이용률 계산 (주차된 자전거 수 / 거치대 총 수)
    df['usage_rate'] = df['parkingBikeTotCnt'] / df['rackTotCnt']

    return df

# 시간대별 데이터 함수
def get_hourly_data():
    """시간대별 자전거 이용률 데이터를 BigQuery에서 가져오는 함수"""
    client = bigquery.Client()

    # 오늘 날짜 기준 시간대별 평균 이용률 쿼리
    query = """
    WITH hourly_data AS (
      SELECT
        EXTRACT(HOUR FROM collection_time) AS hour,
        stationId,
        parkingBikeTotCnt,
        rackTotCnt
      FROM
        `bike-analysis-project-452711.bike_analysis_project.seoul_bike_stations`
      WHERE
        DATE(collection_time) = CURRENT_DATE()
    )

    SELECT
      hour,
      AVG(parkingBikeTotCnt / rackTotCnt) AS usage_rate,
      COUNT(DISTINCT stationId) AS station_count
    FROM
      hourly_data
    GROUP BY
      hour
    ORDER BY
      hour
    """

    # 쿼리 실행 및 DataFrame으로 변환
    hourly_df = client.query(query).to_dataframe()
    return hourly_df

# 지역별 시간대 데이터 함수
def get_district_hourly_data():
    """지역별 시간대별 자전거 이용률 데이터를 가져오는 함수"""
    client = bigquery.Client()

    # 지역 추출을 위한 정규식 패턴 (첫 글자부터 공백 또는 숫자 전까지)
    query = """
    WITH station_districts AS (
      SELECT
        stationId,
        REGEXP_EXTRACT(stationName, r'^([가-힣]+)') AS district
      FROM
        `bike-analysis-project-452711.bike_analysis_project.seoul_bike_stations`
      GROUP BY
        stationId, stationName
    ),

    hourly_data AS (
      SELECT
        EXTRACT(HOUR FROM t.collection_time) AS hour,
        sd.district,
        t.parkingBikeTotCnt,
        t.rackTotCnt
      FROM
        `bike-analysis-project-452711.bike_analysis_project.seoul_bike_stations` t
      JOIN
        station_districts sd
      ON
        t.stationId = sd.stationId
      WHERE
        DATE(t.collection_time) = CURRENT_DATE()
        AND sd.district IS NOT NULL
    )

    SELECT
      hour,
      district,
      AVG(parkingBikeTotCnt / rackTotCnt) AS usage_rate,
      COUNT(DISTINCT stationId) AS station_count
    FROM
      hourly_data
    GROUP BY
      hour, district
    HAVING
      station_count >= 5  -- 최소 5개 이상의 대여소가 있는 지역만 포함
    ORDER BY
      district, hour
    """

    # 쿼리 실행 및 DataFrame으로 변환
    district_hourly_df = client.query(query).to_dataframe()
    return district_hourly_df

# 트렌드 데이터 함수
def get_trend_data(days=7):
    """최근 일주일간의 일별 이용률 트렌드 데이터"""
    client = bigquery.Client()

    query = f"""
    WITH daily_data AS (
      SELECT
        DATE(collection_time) AS date,
        stationId,
        AVG(parkingBikeTotCnt) AS avg_bikes,
        MAX(rackTotCnt) AS rack_count
      FROM
        `bike-analysis-project-452711.bike_analysis_project.seoul_bike_stations`
      WHERE
        DATE(collection_time) BETWEEN DATE_SUB(CURRENT_DATE(), INTERVAL {days} DAY) AND CURRENT_DATE()
      GROUP BY
        date, stationId
    )

    SELECT
      date,
      AVG(avg_bikes / rack_count) AS usage_rate,
      COUNT(DISTINCT stationId) AS station_count
    FROM
      daily_data
    GROUP BY
      date
    ORDER BY
      date
    """

    # 쿼리 실행 및 DataFrame으로 변환
    trend_df = client.query(query).to_dataframe()
    return trend_df

# 데이터 로드
try:
    df = load_data()
except Exception as e:
    st.error(f"데이터 로드 중 오류가 발생했습니다: {str(e)}")
    # 샘플 데이터 생성
    st.warning("샘플 데이터를 대신 표시합니다.")

    # 샘플 데이터 생성
    districts = ['강남', '서초', '송파', '마포', '종로', '성동', '광진', '중구', '용산',
                '서대문', '강서', '영등포', '동작', '관악', '강동', '노원', '도봉', '은평']

    sample_data = []
    for i in range(1000):
        # 지역 이름 + 랜덤 번호로 대여소 이름 생성
        district = districts[i % len(districts)]
        station_name = f"{district} {1000 + i}번 대여소"

        # 거치대 수 (10~50 사이 랜덤)
        rack_cnt = np.random.randint(10, 51)

        # 자전거 수 (0~거치대 수 사이 랜덤)
        bike_cnt = np.random.randint(0, rack_cnt + 1)

        # 현재 시각 정보
        collection_time = datetime.now()
        collection_time_str = collection_time.strftime('%Y-%m-%d %H:%M:%S+09:00')
        collection_hour_kst = collection_time.hour
        collection_minute = collection_time.minute
        collection_date = collection_time.strftime('%Y-%m-%d')

        # 위도/경도 랜덤 생성 (서울 영역 내)
        lat_base = 37.5 + np.random.random() * 0.2
        lon_base = 126.8 + np.random.random() * 0.4

        sample_data.append({
            'stationId': f"ST-{1000 + i}",
            'stationName': station_name,
            'parkingBikeTotCnt': bike_cnt,
            'rackTotCnt': rack_cnt,
            'shared': 0,
            'stationLatitude': lat_base,
            'stationLongitude': lon_base,
            'collection_time': collection_time,
            'collection_time_str': collection_time_str,
            'collection_hour_kst': collection_hour_kst,
            'collection_minute': collection_minute,
            'collection_date': collection_date,
            'usage_rate': bike_cnt / rack_cnt
        })

    df = pd.DataFrame(sample_data)

# 상단 제목 및 설명
st.title('🚲 서울시 공공자전거(따릉이) 현황 대시보드')
st.markdown(f"""
    **마지막 업데이트: {df['collection_time_str'].iloc[0] if not df.empty else datetime.now().strftime('%Y-%m-%d %H:%M:%S')}**

    이 대시보드는 서울시 공공자전거(따릉이) 대여소의 실시간 현황을 보여줍니다.
""")

# 대시보드 메인 지표 (KPI)
st.header('주요 현황')
col1, col2, col3, col4 = st.columns(4)

with col1:
    total_stations = len(df)
    st.metric(label="총 대여소 수", value=f"{total_stations}개")

with col2:
    total_racks = df['rackTotCnt'].sum()
    st.metric(label="총 거치대 수", value=f"{total_racks}대")

with col3:
    total_bikes = df['parkingBikeTotCnt'].sum()
    st.metric(label="주차 자전거 수", value=f"{total_bikes}대")

with col4:
    avg_usage_rate = (df['parkingBikeTotCnt'].sum() / df['rackTotCnt'].sum()) * 100
    st.metric(label="평균 이용률", value=f"{avg_usage_rate:.1f}%")

# 탭 생성
tab1, tab2, tab3 = st.tabs(["📍 대여소 위치 및 현황", "📊 이용률 분석", "⏱️ 시간대별 현황"])

# 탭 1: 대여소 위치 및 현황
with tab1:
    st.subheader("서울시 공공자전거 대여소 위치 및 현황")

    # 필터 설정
    st.sidebar.header("필터 설정")

    # 구별 필터링을 위한 구 이름 추출 (첫 글자부터 공백 또는 숫자 전까지)
    df['district'] = df['stationName'].str.extract(r'^([가-힣]+)')
    districts = sorted(df['district'].dropna().unique())

    selected_district = st.sidebar.selectbox(
        "지역 선택",
        ["전체"] + list(districts)
    )

    # 이용률 범위 필터
    usage_range = st.sidebar.slider(
        "이용률 범위 (%)",
        0, 100, (0, 100)
    )

    # 필터링 적용
    filtered_df = df.copy()
    if selected_district != "전체":
        filtered_df = filtered_df[filtered_df['district'] == selected_district]

    filtered_df = filtered_df[
        (filtered_df['usage_rate'] * 100 >= usage_range[0]) &
        (filtered_df['usage_rate'] * 100 <= usage_range[1])
    ]

    # 지도 시각화
    st.markdown("#### 대여소 위치 지도")
    st.markdown("마커 색상은 자전거 이용률을 나타냅니다 (파란색: 낮음, 빨간색: 높음)")

    # 마커 크기를 주차된 자전거 수에 비례하게 설정
    filtered_df['marker_size'] = filtered_df['parkingBikeTotCnt'] / 2 + 5  # 시각화를 위한 조정

    # 지도 생성
    fig = px.scatter_mapbox(
        filtered_df,
        lat="stationLatitude",
        lon="stationLongitude",
        color="usage_rate",
        size="marker_size",
        hover_name="stationName",
        hover_data={
            "stationId": True,
            "parkingBikeTotCnt": True,
            "rackTotCnt": True,
            "usage_rate": ":.1%",
            "marker_size": False
        },
        color_continuous_scale=px.colors.sequential.Blues_r,  # 낮은 값이 진한 파란색, 높은 값이 빨간색
        zoom=11,
        height=600,
        range_color=[0, 1]
    )

    fig.update_layout(
        mapbox_style="carto-positron",
        margin={"r":0,"t":0,"l":0,"b":0}
    )

    st.plotly_chart(fig, use_container_width=True)

    # 대여소 목록 테이블
    st.markdown("#### 대여소 목록")

    # 테이블에 표시할 컬럼 선택
    display_cols = [
        "stationName", "parkingBikeTotCnt", "rackTotCnt",
        "usage_rate", "collection_hour_kst"
    ]

    # 컬럼명 변경
    renamed_df = filtered_df[display_cols].copy()
    renamed_df.columns = [
        "대여소명", "주차 자전거 수", "거치대 수",
        "이용률", "수집 시간(KST)"
    ]

    # 이용률을 퍼센트로 표시
    renamed_df["이용률"] = (renamed_df["이용률"] * 100).round(1).astype(str) + '%'

    st.dataframe(
        renamed_df.sort_values("주차 자전거 수", ascending=False),
        use_container_width=True
    )

# 탭 2: 이용률 분석
with tab2:
    st.subheader("대여소 이용률 분석")

    # 이용률 히스토그램
    st.markdown("#### 이용률 분포")

    fig = px.histogram(
        df,
        x="usage_rate",
        nbins=20,
        labels={"usage_rate": "이용률"},
        color_discrete_sequence=['#1E90FF'],
        title="대여소 이용률 분포"
    )

    fig.update_layout(
        xaxis_title="이용률",
        yaxis_title="대여소 수",
        xaxis=dict(tickformat=".0%")
    )

    st.plotly_chart(fig, use_container_width=True)

    # 이용률 상위/하위 대여소
    col1, col2 = st.columns(2)

    with col1:
        st.markdown("#### 이용률 상위 10개 대여소")
        top_usage = df.sort_values("usage_rate", ascending=False).head(10)

        fig = px.bar(
            top_usage,
            y="stationName",
            x="usage_rate",
            orientation='h',
            color="usage_rate",
            color_continuous_scale=px.colors.sequential.Reds,
            labels={"stationName": "대여소명", "usage_rate": "이용률"},
            text_auto='.1%'
        )

        fig.update_layout(
            xaxis=dict(tickformat=".0%"),
            height=400
        )

        st.plotly_chart(fig, use_container_width=True)

    with col2:
        st.markdown("#### 이용률 하위 10개 대여소")
        # 이용률이 0인 대여소는 제외하고 낮은 순으로 정렬
        bottom_usage = df[df['usage_rate'] > 0].sort_values("usage_rate").head(10)

        fig = px.bar(
            bottom_usage,
            y="stationName",
            x="usage_rate",
            orientation='h',
            color="usage_rate",
            color_continuous_scale=px.colors.sequential.Blues,
            labels={"stationName": "대여소명", "usage_rate": "이용률"},
            text_auto='.1%'
        )

        fig.update_layout(
            xaxis=dict(tickformat=".0%"),
            height=400
        )

        st.plotly_chart(fig, use_container_width=True)

    # 거치대 수별 이용률 산점도
    st.markdown("#### 거치대 수와 이용률 관계")

    fig = px.scatter(
        df,
        x="rackTotCnt",
        y="usage_rate",
        color="parkingBikeTotCnt",
        size="parkingBikeTotCnt",
        hover_name="stationName",
        color_continuous_scale=px.colors.sequential.Viridis,
        labels={
            "rackTotCnt": "거치대 수",
            "usage_rate": "이용률",
            "parkingBikeTotCnt": "현재 자전거 수"
        }
    )

    fig.update_layout(
        yaxis=dict(tickformat=".0%"),
        height=500
    )

    st.plotly_chart(fig, use_container_width=True)

# 탭 3: 시간대별 현황
with tab3:
    st.subheader("시간대별 자전거 이용 현황")

    # 시간별 데이터 로드 (BigQuery로부터)
    try:
        with st.spinner("시간대별 데이터를 불러오는 중..."):
            hourly_data = get_hourly_data()

            # 데이터가 없는 시간대에 대해 0 값으로 채우기
            all_hours = pd.DataFrame({"hour": range(24)})
            hourly_data = pd.merge(all_hours, hourly_data, on="hour", how="left").fillna(0)

            # 시간대별 이용률 그래프
            st.markdown("#### 시간대별 평균 이용률")

            fig = px.line(
                hourly_data,
                x="hour",
                y="usage_rate",
                labels={"hour": "시간 (KST)", "usage_rate": "평균 이용률"},
                markers=True,
                line_shape="spline"
            )

            fig.update_layout(
                xaxis=dict(
                    tickmode='array',
                    tickvals=list(range(0, 24)),
                    ticktext=[f"{h}시" for h in range(0, 24)]
                ),
                yaxis=dict(tickformat=".0%"),
                height=400
            )

            st.plotly_chart(fig, use_container_width=True)
    except Exception as e:
        st.error(f"시간대별 데이터 로드 중 오류가 발생했습니다: {str(e)}")
        st.info("샘플 데이터를 대신 표시합니다.")

        # 샘플 데이터 생성 (오류 발생 시 대체용)
        sample_hours = list(range(0, 24))
        usage_by_hour = [
            0.3, 0.2, 0.15, 0.1, 0.2, 0.4, 0.6, 0.8, 0.7, 0.5,
            0.4, 0.5, 0.6, 0.5, 0.4, 0.5, 0.7, 0.8, 0.7, 0.5, 0.4, 0.3, 0.2, 0.1
        ]

        hourly_data = pd.DataFrame({
            "hour": sample_hours,
            "usage_rate": usage_by_hour
        })

        # 샘플 시간대별 이용률 그래프
        st.markdown("#### 시간대별 평균 이용률 (샘플 데이터)")

        fig = px.line(
            hourly_data,
            x="hour",
            y="usage_rate",
            labels={"hour": "시간 (KST)", "usage_rate": "평균 이용률"},
            markers=True,
            line_shape="spline"
        )

        fig.update_layout(
            xaxis=dict(
                tickmode='array',
                tickvals=list(range(0, 24)),
                ticktext=[f"{h}시" for h in range(0, 24)]
            ),
            yaxis=dict(tickformat=".0%"),
            height=400
        )

        st.plotly_chart(fig, use_container_width=True)

    # 지역별 이용 패턴
    st.markdown("#### 주요 지역별 이용 패턴")

    try:
        with st.spinner("지역별 시간대 데이터를 불러오는 중..."):
            district_df = get_district_hourly_data()

            # 주요 지역만 선택 (데이터 포인트가 많은 상위 5개 지역)
            top_districts = district_df.groupby('district')['station_count'].mean().nlargest(5).index.tolist()
            district_df = district_df[district_df['district'].isin(top_districts)]

            fig = px.line(
                district_df,
                x="hour",
                y="usage_rate",
                color="district",
                labels={"hour": "시간 (KST)", "usage_rate": "평균 이용률", "district": "지역"},
                line_shape="spline"
            )

            fig.update_layout(
                xaxis=dict(
                    tickmode='array',
                    tickvals=list(range(0, 24)),
                    ticktext=[f"{h}시" for h in range(0, 24)]
                ),
                yaxis=dict(tickformat=".0%"),
                height=500
            )

            st.plotly_chart(fig, use_container_width=True)
    except Exception as e:
        st.error(f"지역별 시간대 데이터 로드 중 오류가 발생했습니다: {str(e)}")
        st.info("샘플 데이터를 대신 표시합니다.")

        # 가상의 지역별 이용 패턴 데이터 생성 (오류 발생 시 대체용)
        districts = ["강남", "서초", "송파", "마포", "종로"]
        district_data = []

        for district in districts:
            base = hourly_data["usage_rate"].values.copy()
            if district == "강남":
                for i in [7, 8, 18, 19]:
                    if i < len(base):
                        base[i] *= 1.2
            elif district == "마포":
                for i in range(18, 22):
                    if i < len(base):
                        base[i] *= 1.15
            elif district == "종로":
                for i in range(11, 14):
                    if i < len(base):
                        base[i] *= 1.25
            sample_hours = list(range(0, 24))
            for hour, rate in zip(sample_hours, base):
                district_data.append({
                    "district": district,
                    "hour": hour,
                    "usage_rate": rate
                })

        district_df = pd.DataFrame(district_data)

        fig = px.line(
            district_df,
            x="hour",
            y="usage_rate",
            color="district",
            labels={"hour": "시간 (KST)", "usage_rate": "평균 이용률", "district": "지역"},
            line_shape="spline"
        )

        fig.update_layout(
            xaxis=dict(
                tickmode='array',
                tickvals=list(range(0, 24)),
                ticktext=[f"{h}시" for h in range(0, 24)]
            ),
            yaxis=dict(tickformat=".0%"),
            height=500
        )

        st.plotly_chart(fig, use_container_width=True)

    # 주간 트렌드 차트 추가
    st.markdown("#### 일주일간 이용률 트렌드")

    try:
        with st.spinner("주간 트렌드 데이터를 불러오는 중..."):
            trend_data = get_trend_data(days=7)

            fig = px.line(
                trend_data,
                x="date",
                y="usage_rate",
                labels={"date": "날짜", "usage_rate": "평균 이용률"},
                markers=True,
                line_shape="spline"
            )

            fig.update_layout(
                yaxis=dict(tickformat=".0%"),
                height=400
            )

            st.plotly_chart(fig, use_container_width=True)
    except Exception as e:
        st.error(f"주간 트렌드 데이터 로드 중 오류가 발생했습니다: {str(e)}")
        st.info("주간 트렌드 데이터를 표시할 수 없습니다.")

# 사이드바 추가 정보
st.sidebar.markdown("---")
st.sidebar.markdown("### 대시보드 정보")
st.sidebar.info(
    """
    **서울시 공공자전거(따릉이) 현황 대시보드**

    이 대시보드는 서울시 공공자전거 실시간 대여소 현황을 시각화합니다.

    데이터 출처: 서울시 공공자전거 BigQuery 데이터
    """
)
```
