2. **Î∂àÏùºÏπòÌï† Í≤ΩÏö∞:**  
   - ÏÇ¨Ïö©ÏûêÎ≥Ñ Î∞òÏùë(ÌÅ¥Î¶≠, ÏßÄÏ∂ú)Ïù¥ Îã§Î•¥Í±∞ÎÇò, Ïã§Ï†úÎ°ú Îã§Î•∏ Ï∫†ÌéòÏù∏Ïùº Í∞ÄÎä•ÏÑ±.




```python
# Í¥ëÍ≥† Ï∫†ÌéòÏù∏ Í∏∞Ï§ÄÏúºÎ°ú Í∑∏Î£πÌôîÌïòÍ≥†, Í¥ëÍ≥† ÏÑ±Í≥º Î≥ÄÏàòÏùò ÏùºÏπò Ïó¨Î∂Ä Î∂ÑÏÑù
# Í∑∏Î£πÌôî Í∏∞Ï§Ä: Í¥ëÍ≥†_Ï±ÑÎÑê, Í¥ëÍ≥†_ÌòïÏãù, Í¥ëÍ≥†_ÌÉÄÍ≤üÌåÖ, Í¥ëÍ≥†_ÏãúÏûëÏùº, Í¥ëÍ≥†_Ï¢ÖÎ£åÏùº
campaign_group = ad_data.groupby(['Í¥ëÍ≥†_Ï±ÑÎÑê', 'Í¥ëÍ≥†_ÌòïÏãù', 'Í¥ëÍ≥†_ÌÉÄÍ≤üÌåÖ', 'Í¥ëÍ≥†_ÏãúÏûëÏùº', 'Í¥ëÍ≥†_Ï¢ÖÎ£åÏùº'])

# ÏÑ±Í≥º ÏßÄÌëú(Í¥ëÍ≥†_ÎÖ∏Ï∂ú_ÌöüÏàò, Í¥ëÍ≥†_ÌÅ¥Î¶≠_ÌöüÏàò, Í¥ëÍ≥†_ÏßÄÏ∂úÏï°)Ïóê ÎåÄÌï¥ ÏµúÏÜåÍ∞íÍ≥º ÏµúÎåÄÍ∞íÏùÑ ÎπÑÍµêÌïòÏó¨ ÏùºÏπò Ïó¨Î∂Ä ÌôïÏù∏
result = campaign_group.agg({
    'Í¥ëÍ≥†_ÎÖ∏Ï∂ú_ÌöüÏàò': ['min', 'max', 'std'],
    'Í¥ëÍ≥†_ÌÅ¥Î¶≠_ÌöüÏàò': ['min', 'max', 'std'],
    'Í¥ëÍ≥†_ÏßÄÏ∂úÏï°': ['min', 'max', 'std']
})

# ÏùºÏπò Ïó¨Î∂ÄÎ•º ÏâΩÍ≤å ÌôïÏù∏Ìï† Ïàò ÏûàÎèÑÎ°ù Ï∂îÍ∞Ä Ïª¨Îüº ÏÉùÏÑ±
result['ÎÖ∏Ï∂ú_ÏùºÏπò'] = result[('Í¥ëÍ≥†_ÎÖ∏Ï∂ú_ÌöüÏàò', 'min')] == result[('Í¥ëÍ≥†_ÎÖ∏Ï∂ú_ÌöüÏàò', 'max')]
result['ÌÅ¥Î¶≠_ÏùºÏπò'] = result[('Í¥ëÍ≥†_ÌÅ¥Î¶≠_ÌöüÏàò', 'min')] == result[('Í¥ëÍ≥†_ÌÅ¥Î¶≠_ÌöüÏàò', 'max')]
result['ÏßÄÏ∂ú_ÏùºÏπò'] = result[('Í¥ëÍ≥†_ÏßÄÏ∂úÏï°', 'min')] == result[('Í¥ëÍ≥†_ÏßÄÏ∂úÏï°', 'max')]

# ÏùºÏπò Ïó¨Î∂ÄÏôÄ ÌëúÏ§ÄÌé∏Ï∞®Î•º Ìï®Íªò ÌôïÏù∏

# ÏµúÏ¢Ö Í≤∞Í≥º Î∞òÌôò
result[['ÎÖ∏Ï∂ú_ÏùºÏπò', 'ÌÅ¥Î¶≠_ÏùºÏπò', 'ÏßÄÏ∂ú_ÏùºÏπò']].value_counts()
```

#### **1. ÏÑ±Í≥º ÏßÄÌëú (Performance Metrics)**
   - **Í¥ëÍ≥† ÌÅ¥Î¶≠Î•† (CTR)**: `Í¥ëÍ≥†_ÌÅ¥Î¶≠_ÌöüÏàò / Í¥ëÍ≥†_ÎÖ∏Ï∂ú_ÌöüÏàò * 100`
     - Í¥ëÍ≥†Í∞Ä ÎÖ∏Ï∂úÎêú ÎåÄÎπÑ Ïã§Ï†úÎ°ú ÌÅ¥Î¶≠Îêú ÎπÑÏú®. ÎÜíÏùÄ CTRÏùÄ Í¥ëÍ≥†Ïùò Í¥ÄÏã¨ÎèÑÎ•º ÏùòÎØ∏.
   - **Ï†ÑÌôòÏú® (CVR)**: `Ï†ÑÌôòÏú®` (Ïù¥ÎØ∏ Îç∞Ïù¥ÌÑ∞Ïóê Ï°¥Ïû¨)
     - ÌÅ¥Î¶≠ ÌõÑ Ïã§Ï†ú Íµ¨Îß§ÎÇò ÏõêÌïòÎäî Ïï°ÏÖòÏù¥ ÏùºÏñ¥ÎÇú ÎπÑÏú®. ÏÑ±Í≥ºÏùò Í∞ÄÏû• Ï§ëÏöîÌïú ÏßÄÌëú.
   - **Îß§Ï∂úÏï° (Revenue)**: `Îß§Ï∂úÏï°` (Ïù¥ÎØ∏ Îç∞Ïù¥ÌÑ∞Ïóê Ï°¥Ïû¨)

   - **Í¥ëÍ≥† ÌöåÏÉÅÎ•† (Ad Recall Rate)**: `Í¥ëÍ≥†_ÌöåÏÉÅÎ•†` (Ïù¥ÎØ∏ Îç∞Ïù¥ÌÑ∞Ïóê Ï°¥Ïû¨)

   - **Í¥ëÍ≥† ÌÉúÎèÑ Î≥ÄÌôî (Ad Attitude Shift)**: `Í¥ëÍ≥†_ÌÉúÎèÑ_Î≥ÄÌôî` (Ïù¥ÎØ∏ Îç∞Ïù¥ÌÑ∞Ïóê Ï°¥Ïû¨)
     - Í¥ëÍ≥† ÎÖ∏Ï∂ú Ï†ÑÌõÑÏùò Î∏åÎûúÎìúÎÇò Ï†úÌíàÏóê ÎåÄÌïú Ïù∏Ïãù Î≥ÄÌôî.

#### **2. Ìö®Ïú®ÏÑ± ÏßÄÌëú (Efficiency Metrics)**
   - **ROAS (Í¥ëÍ≥† ÏàòÏùµÎ•†)**: `Îß§Ï∂úÏï° / Í¥ëÍ≥†_ÏßÄÏ∂úÏï°`
     - Í¥ëÍ≥†ÎπÑ 1ÏõêÎãπ Î∞úÏÉùÌïú ÏàòÏùµ. ROASÍ∞Ä 1 Ïù¥ÏÉÅÏù¥Î©¥ ÎπÑÏö© ÎåÄÎπÑ ÏàòÏùµÏù¥ Í∏çÏ†ïÏ†Å.
   - **CPA (Ï†ÑÌôòÎãπ ÎπÑÏö©)**: `Í¥ëÍ≥†_ÏßÄÏ∂úÏï° / (Ï†ÑÌôòÏú® * ÏõπÏÇ¨Ïù¥Ìä∏_Î∞©Î¨∏_Ïàò)`

   - **CPC (ÌÅ¥Î¶≠Îãπ ÎπÑÏö©)**: `Í¥ëÍ≥†_ÏßÄÏ∂úÏï° / Í¥ëÍ≥†_ÌÅ¥Î¶≠_ÌöüÏàò`
     - ÌÅ¥Î¶≠ 1ÌöåÎãπ Î∞úÏÉùÌïú ÎπÑÏö©. Í¥ëÍ≥† ÏßëÌñâ ÎπÑÏö© Í¥ÄÎ¶¨Ïóê Ï§ëÏöî.
   - **CPM (ÎÖ∏Ï∂ú 1,000ÌöåÎãπ ÎπÑÏö©)**: `Í¥ëÍ≥†_ÏßÄÏ∂úÏï° / Í¥ëÍ≥†_ÎÖ∏Ï∂ú_ÌöüÏàò * 1000`
     - Í¥ëÍ≥† ÎÖ∏Ï∂ú Í∏∞Ï§Ä ÎπÑÏö©. ÎÖ∏Ï∂ú Ï§ëÏã¨Ïùò Í¥ëÍ≥†ÏóêÏÑúÎäî CPMÏùÑ ÎÇÆÏ∂îÎäî Í≤ÉÏù¥ Ï§ëÏöî.

#### **3. ÏÇ¨Ïö©Ïûê Î∞òÏùë ÏßÄÌëú (User Engagement Metrics)**
   - **ÏõπÏÇ¨Ïù¥Ìä∏ Î∞©Î¨∏Ïú® (Visit Rate)**: `ÏõπÏÇ¨Ïù¥Ìä∏_Î∞©Î¨∏_Ïàò / Í¥ëÍ≥†_ÌÅ¥Î¶≠_ÌöüÏàò * 100`
     - Í¥ëÍ≥† ÌÅ¥Î¶≠ ÌõÑ Ïã§Ï†ú ÏõπÏÇ¨Ïù¥Ìä∏Î°ú Ïú†ÏûÖÎêú ÎπÑÏú®.
   - **ÌéòÏù¥ÏßÄÎ∑∞Îãπ Ï≤¥Î•òÏú® (Page Views per Visit)**: `ÌéòÏù¥ÏßÄÎ∑∞ / ÏõπÏÇ¨Ïù¥Ìä∏_Î∞©Î¨∏_Ïàò`
     - Î∞©Î¨∏Îãπ ÌèâÍ∑† ÌéòÏù¥ÏßÄ Ï°∞Ìöå Ïàò. ÎÜíÏùÄ Í∞íÏùÄ ÏΩòÌÖêÏ∏†Ïóê ÎåÄÌïú Í¥ÄÏã¨ÎèÑÎ•º ÏùòÎØ∏.

#### **4. ÎπÑÏö© ÏµúÏ†ÅÌôî ÏßÄÌëú (Cost Optimization Metrics)**
   - **Í¥ëÍ≥† ÎπÑÏö© ÎπÑÏú®:** `Í¥ëÍ≥†_ÏßÄÏ∂úÏï° / Îß§Ï∂úÏï° * 100`
     - ÏàòÏùµ ÎåÄÎπÑ Í¥ëÍ≥†ÎπÑ ÎπÑÏú®. 30% Ïù¥ÌïòÍ∞Ä Ïù¥ÏÉÅÏ†Å.
   - **Ï±ÑÎÑêÎ≥Ñ CPA:** Ï±ÑÎÑêÎ≥Ñ Ï†ÑÌôòÎãπ ÎπÑÏö© ÎπÑÍµê.

---


```python
# ROAS, CTR, CPA Í≥ÑÏÇ∞
matched_data['CTR'] = (matched_data['Í¥ëÍ≥†_ÌÅ¥Î¶≠_ÌöüÏàò'] / matched_data['Í¥ëÍ≥†_ÎÖ∏Ï∂ú_ÌöüÏàò']) * 100
matched_data['ROAS'] = matched_data['Îß§Ï∂úÏï°'] / matched_data['Í¥ëÍ≥†_ÏßÄÏ∂úÏï°']
matched_data['CPA'] = matched_data['Í¥ëÍ≥†_ÏßÄÏ∂úÏï°'] / (matched_data['Ï†ÑÌôòÏú®'] * matched_data['ÏõπÏÇ¨Ïù¥Ìä∏_Î∞©Î¨∏_Ïàò'])

# Ï∫†ÌéòÏù∏ Í∏∞Ï§Ä ÏÑ±Í≥º Î∂ÑÏÑù
campaign_performance = matched_data.groupby(['Í¥ëÍ≥†_Ï±ÑÎÑê', 'Í¥ëÍ≥†_ÌòïÏãù', 'Í¥ëÍ≥†_ÌÉÄÍ≤üÌåÖ', 'Í¥ëÍ≥†_ÏãúÏûëÏùº', 'Í¥ëÍ≥†_Ï¢ÖÎ£åÏùº']).agg({
    'Í¥ëÍ≥†_ÎÖ∏Ï∂ú_ÌöüÏàò': 'sum',
    'Í¥ëÍ≥†_ÌÅ¥Î¶≠_ÌöüÏàò': 'sum',
    'Í¥ëÍ≥†_ÏßÄÏ∂úÏï°': 'sum',
    'Îß§Ï∂úÏï°': 'sum',
    'CTR': 'mean',
    'ROAS': 'mean',
    'CPA': 'mean'
}).reset_index()

# Í≤∞Í≥º ÌÖåÏù¥Î∏î ÌëúÏãú
campaign_performance
```
